#### **I will first handle errors and then connect my table to sql database.**

```python
import pyodbc
import pandas as pd
import numpy as np

# Step 1 — Clean dtypes
df_clean = df.copy()
df_clean["review_rating"] = pd.to_numeric(df_clean["review_rating"], errors="coerce").astype(float)

# Convert ALL non-numeric columns to str explicitly
for col in df_clean.columns:
    if df_clean[col].dtype.name == "category" or df_clean[col].dtype == "object":
        df_clean[col] = df_clean[col].astype(str)

print("Dtypes after fix:")
print(df_clean.dtypes)

# Step 2 — Build rows with EXPLICIT per-column casting (safest method)
def cast_value(val, dtype):
    if pd.isna(val) if not isinstance(val, str) else val in ("nan", "None", ""):
        return None
    if "int" in str(dtype):   return int(val)
    if "float" in str(dtype): return float(val)
    return str(val)

rows = []
for _, row in df_clean.iterrows():
    rows.append(tuple(
        cast_value(row[col], df_clean[col].dtype) for col in df_clean.columns
    ))

# Verify first row looks correct
print("\nFirst row sample:")
for col, val in zip(df_clean.columns, rows[0]):
    print(f"  {col}: {repr(val)} ({type(val).__name__})")
```

```python
# Step 3 — Connect and insert
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=127.0.0.1,1433;"
    "DATABASE=cust_behaviour;"
    "UID=ADNAN;"
    "PWD=Oracle@123;"
    "TrustServerCertificate=yes;"
)
cursor = conn.cursor()
table_name = "customer"

# Drop and recreate
cursor.execute(f"IF OBJECT_ID('{table_name}', 'U') IS NOT NULL DROP TABLE {table_name}")
conn.commit()

def pandas_type_to_sql(dtype):
    if "int"   in str(dtype): return "INT"
    if "float" in str(dtype): return "FLOAT"
    return "NVARCHAR(255)"

cols = ", ".join(f"[{col}] {pandas_type_to_sql(df_clean[col].dtype)}" for col in df_clean.columns)
cursor.execute(f"CREATE TABLE {table_name} ({cols})")
conn.commit()
print(f"\n✅ Table '{table_name}' created!")

# Insert
chunk_size = 500
placeholders = ", ".join(["?"] * len(df_clean.columns))
insert_sql = f"INSERT INTO {table_name} VALUES ({placeholders})"

for i in range(0, len(rows), chunk_size):
    chunk = rows[i:i+chunk_size]
    cursor.executemany(insert_sql, chunk)
    conn.commit()
    print(f"  ✅ Inserted rows {i+1}–{min(i+chunk_size, len(rows))}")

print(f"\n✅ All {len(df_clean)} rows loaded into '{table_name}' successfully!")
cursor.close()
conn.close()
```
